````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os, sys
from scipy import signal, interpolate, ndimage
````

````python
input_dir = '/home/jovyan/data/hw1_data/'
all_filenames = np.asarray(os.listdir(input_dir))
good_filenames = np.asarray([s.startswith('im2-') for s in all_filenames])
filenames_in = all_filenames[good_filenames]
sort_indices = np.argsort(np.asarray([int(s.split('-')[1].split('.')[0]) for s in filenames_in]))
filenames_in = filenames_in[sort_indices]
````

````python
def load_and_resize(fname, size_multiplier = 3):
    '''
    takes in a fname for a frame of the video,
    resizes it to given size,
    returns as np array
    '''
    array_in = Image.open(fname).convert('L')
    newsize = (size_multiplier*array_in.size[0], size_multiplier*array_in.size[1])
    array_big = array_in.resize(newsize, resample = Image.NEAREST)
    
    # remove the empty edges - they will screw up the convolution
    # all those edges are where the array is identically equal to 255
    plt.imshow(np.asarray(array_big) == 255)
    plt.show()
    
    return np.asarray(array_big)
````

````python
def compute_shift(frame, frame0):
    '''
    take in two frames, returns the optimal x and y distance to move them to stack on top of each other
    '''
    convolution = signal.fftconvolve(frame, frame0, mode='same')
    #plt.imshow(convolution)
    #plt.show()
    
    # need a way to find the max correlation at sub px level
    # try an interpolation function
    xx = np.arange(0, frame.shape[0], 1)
    yy = np.arange(0, frame.shape[1], 1)
    interpf = interpolate.RectBivariateSpline(xx, yy, convolution)
    # but the problem is how to find x,y of the maximum without just testing lots of points
    
    # try supersampling a small box around maximum of convolution
    max_idx = np.argwhere(convolution == np.max(convolution))[0]
    conv_center = convolution[max_idx[0]-2:max_idx[0]+2, max_idx[1]-2:max_idx[1]+2]
    #linspace includes endpoint. this is 5 pixels supersampled by factor of 100
    xx_hires = np.linspace(max_idx[0]-2, max_idx[0]+2, 500) 
    yy_hires = np.linspace(max_idx[1]-2, max_idx[1]+2, 500)
    hi_res_center = interpf(xx_hires, yy_hires)
    #plt.imshow(hi_res_center)
    #plt.show()
    center_idx = np.argwhere(hi_res_center == np.max(hi_res_center))[0]
    #print(center_idx)
    (dx, dy) = (center_idx / 100.) + (max_idx[0]-2, max_idx[1]-2) - (frame0.shape[0]/2, frame0.shape[1]/2)
    
    # this gives us an answer, who knows if it is any good
    # some things to test: 
    # how does the answer change if we change the order of the spline interpolation to something other than cubic?
    # how does the answer change if we use an actually-intelligent fitting function instead of a spline interpolation?
    
    return dx, dy
    
````

````python
def apply_shift(frame, dx, dy):
    '''
    applies a sub-pixel shift to an image
    '''
    shifted_frame = ndimage.interpolation.shift(frame, (dx, dy))
    
    return shifted_frame
    
````

````python
shifted_frames = []
size_multiplier = 3
frame0 = load_and_resize(input_dir + filenames_in[0], size_multiplier = size_multiplier)
#for i in range(1,len(filenames_in)):
for i in range(1,4):
    fname_in = input_dir + filenames_in[i]
    frame = load_and_resize(fname_in, size_multiplier = size_multiplier)
    plt.imshow(frame)
    plt.show()
    dx,dy = compute_shift(frame, frame0)
    print(f"File {filenames_in[i]} requires a shift of {dx, dy}")
    
    shifted_frame = apply_shift(frame, dx, dy)
    #plt.imshow(shifted_frame)
    #plt.show()
    shifted_frames.append(shifted_frame)
````

````python
shifted_frames = np.asarray(shifted_frames)
summed_frame = np.sum(shifted_frames, axis = 0)
plt.imshow(summed_frame)
plt.show()
````

# #2: An elementary introduction to spectral audio compression

In this problem, we'll explore the very basics of audio compression in the spectral domain using numpy and scipy. We'll do a bit of visualization with matplotlib, but since that is covered later in the course, we'll provide those functions for you.

Audio compression is a large and complex topic, and the design of a format for compressed audio such as the popular [MP3](http://en.wikipedia.org/wiki/MP3) is too complex to cover in detail here. However, we will introduce the basic tools that most such compression formats use, namely:

1. Converting the input signal to the frequency domain by taking a Fast Fourier Transform (FFT).

2. Dropping information in the frequency domain, resulting in a smaller amount of data.

3. Reconstructing back the signal in the time domain from this smaller representation of the signal.

Steps 1 and 2 above are the 'encoding' part of signal compression, and step 3 is the 'decoding' part. For this reason, the tools that perform these steps are typically referred to as signal 'codecs', short for encoders/decoders.

Note that here we say 'signal': while MP3 is an audio format, the same ideas apply to the compression of digital images with formats such as JPEG and video.  Virtually all multimedia technologies we use today, from audio players to cell phones, digital cameras and YouTubeVideo, are based on sophisticated extensions and applications of these simple ideas.

Let's first load the plotting tools and importing some tools we'll need later:

We define a simple utility function to listen to audio files right in the browser:

````python
def Audio(fname):
    """Provide a player widget for an audio file.
    
    Parameters
    ==========
    fname : string
      Filename to be played.
      
    Warning
    =======
    
    Browsers cache audio very aggressively. If you change an
    audio file on disk and are trying to listen to the  new version, you 
    may want to 
    """
    from IPython.display import HTML, display
    
    # Find out file extension and deduce MIME type for audio format
    ext = os.path.splitext(fname)[1].replace('.', '').lower()
    mimetype = 'audio/' + ('mpeg' if ext == 'mp3' else ext)
    
    tpl = """<p>{fname}:</p>
<audio controls>
    <source src="files/{fname}" type="{mimetype}">

Your browser does not support the Audio element; you can play 
<a href="files/{fname}">this file</a> manually.
</audio>
"""
    display(HTML(tpl.format(**locals())))
````

We also define a convenience wrapper around `plt.specgram`, [matplotlib's spectrogram function](http://matplotlib.org/api/mlab_api.html#matplotlib.mlab.specgram), with a colorbar and control over the color limits displayed. This will make it easier to compare across different signals with the same colors for all inputs.

````python
def specgram_cbar(x, title=None, clim=(0, 80) ):
    """Plot spectrogram with a colorbar and range normalization.
    
    Call matplotlib's specgram function, with a custom figure size, 
    automatic colobar, title and custom color limits to ease 
    comparison across multiple figures.
    
    Parameters
    ==========
    x : array
      One-dimensional array whose spectrogram should be plotted.
      
    title : string
      Optional title for the figure.
      
    clim : 2-tuple
      Range for the color limits plotted in the spectrogram.
    """
    f = plt.figure(figsize=(10,3))
    plt.specgram(x, Fs=2)
    plt.colorbar()
    plt.clim(*clim)
    if title is not None:
        plt.title(title)
    plt.show()
````
