**Caption**: The colored contours show the ammonia concentration in parts per million inverted from nadir brightness temperatures during PJ1 flyby assuming that the deep water abundance is 0.06% (0.65 times solar). b) In your own words, what story is this figure trying to convey? What does it do well? What could have been done better? What elements didn't need to be present to still convey the same story?

This figure shows the ammonia abundance in Jupiter's troposphere as derived from the Juno microwave radiometer (MWR) instrument. The figure is trying to tell several different stories: It shows the strong global ammonia depletion above $\sim$10 bar as well as the exception to that rule in the equatorial zone (EZ). It shows a strange and unexplained inversion in the ammonia abundance just below the ammonia cloud condensation level at 0.7 bar pressures. Finally, it shows the rapid drop to zero abundance above the ammonia cloud at 0.7 bar.

The plot does a good job in the following ways:
- It clearly shows the contrast between the EZ and the rest of the atmosophere
- Similarly, the cloud condensation level can be clearly seen in the plot
- The choice to use contours instead of smooth colors makes it easy to differentiate gradual vs sharp transitions in ammonia abundance and to follow contours of equal abundance
- The choice to use a logarithmic y-axis is a good one

The plot has the following weaknesses:
- There is no colorbar label
- The colormap is not perceptually uniform and likely challenging for people with colorblindness
- The colormap is diverging for no reason - white does not represent anything physically important
- The contours came out a bit angular, which is a function of the sample density, but this is not made clear in any way
- The theoretical ammonia, NH$_4$SH, and water cloud condensation pressures, which predict which regions of the atmosphere may contain a jump in ammonia abundance, are absent from the figure
- There is no indication of the uncertainty in these measurements; is the inversion just below the condensation layer even real?

I think every element needs to be present to convey the story, but more elements need to be there, too!  My suggested improvements are:
- Use a perceptually-uniform, non-diverging, colorblind-friendly, labeled colormap
- Label theoretical cloud condensation pressures
- Label the locations of zones and belts in the atmosphere, especially the Equatorial Zone and Southern Equatorial Belt, since it looks like interesting stuff goes on there
- It's unclear how to incorporate the uncertainties here, but I feel they should at least be mentioned in the figure caption, something like "the typical uncertainty in ammonia abundance is xx ppm" or "xx percent." My labmate is showing that the map you get really depends strongly on your assumed parameters, so an even better idea would be to make an interactive plot where you can slide different assumed parameters within their uncertainties and see how it affects this map

## Reproduce one of your old published-paper quality plots with Bokeh or matplotlib.

a) Provide the original plot, the recreated Bokeh/matlibplot plot, and the Python code used to make the new plot.  You can ask us for an example from us if you don't have your own plots.

<img src="https://cfn-live-content-bucket-iop-org.s3.amazonaws.com/journals/2632-3338/2/1/3/1/psjabc48af8_hr.jpg?AWSAccessKeyId=AKIAYDKQL6LTV7YY2HIK&Expires=1644865350&Signature=x6I5LK6%2FD06HKAb7uXsdnZw8APs%3D">

**Caption**: Effect of changing radiative transfer model parameters on the radio spectrum of Uranus. In each panel, one parameter was changed from its nominal value; the resulting model spectrum was subtracted from the nominal model such that ΔTB represents the departure from the nominal model. Legend labels denote a multiplicative factor applied to the model parameter of interest, where 1.0 is the nominal value (30× solar for H2S and PH3; 1× solar for NH3, 1.0 for H2S hrel). The black points plotted along the zero line show the size of the error bars on the data at the observed frequencies. All models were produced assuming viewing geometry from the sub-observer point. 

From [Molter et al. (2021), PSJ](https://iopscience.iop.org/article/10.3847/PSJ/abc48a)

```python
## re-make at least one of the panels using Bokeh
import numpy as np
from bokeh.plotting import figure, show, output_file, output_notebook
from bokeh import palettes
from bokeh import models
import glob
# show inline in the notebook
output_notebook()

TOOLS="hover,crosshair,pan,wheel_zoom,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select"
p = figure(tools=TOOLS,title="Top Title with Toolbar", toolbar_location="above",
           plot_width=600, plot_height=600,
          x_axis_type="log")

# import the real data and plot. needs special Whisker function for errorbars
[freq, tb, tberr] = np.load('data/real_data.npy')
upper = [x+e for x,e in zip(tb, tberr) ]
lower = [x-e for x,e in zip(tb, tberr) ]
source = models.ColumnDataSource(data=dict(freq=freq, tb=tb, upper=upper, lower=lower))
p.scatter(freq, tb, line_color = 'black', line_width=5)
w = models.Whisker(source=source, base="freq", upper="upper", lower="lower", level="overlay", line_width = 2)
w.lower_head.line_width = 2 #this was annoying to figure out how to do
w.upper_head.line_width = 2
p.add_layout(w)

# import the models and plot those
infnames = np.asarray(glob.glob("data/nh3*.npy"))
abus = np.asarray([float(fname.split('.')[0].split('_')[1]) for fname in infnames])
sort_idx = np.argsort(abus)
abus = abus[sort_idx]
infnames = infnames[sort_idx]
abus_use = np.asarray([1e-2, 1e-1, 3e-1, 3e0, 1e1])
infnames_use = infnames[np.isin(abus,abus_use)]

colors = palettes.plasma(len(infnames_use))
for i in range(len(infnames_use)):
    fname = infnames_use[i]
    print(fname)
    label = fname.split('.')[0].split('_')[1]
    [freq, tb] = np.load(fname)

    p.line(freq, tb, line_width=4, line_color = colors[i], legend_label=label)

# beautify
p.xaxis.axis_label = 'Frequency (GHz)'
p.yaxis.axis_label = 'Normalized Brightness Temperature (K)'

p.axis.axis_label_text_font_style = 'normal'
p.axis.axis_label_text_font_size = '20pt'
p.axis.axis_label_text_font_size = '18pt'
p.axis.major_label_text_font_size = '18pt'

p.legend.location = 'bottom_right'
p.legend.label_text_font_size = '14pt'

output_file("uranus_radio_spectra.html", title="Uranus Radio Spectra")
show(p)  # open a browser
```

##  Volcano-induced Blastwave moving through California in 2022

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/f9/Shockwave_from_Hunga_Tonga_GOES_17.gif/440px-Shockwave_from_Hunga_Tonga_GOES_17.gif" width="30%">

The massive volcanic eruption near Tonga in the Pacific Ocean (20.550°S 175.385°W) occurred at 04:14:45 [UTC](https://en.wikipedia.org/wiki/Coordinated_Universal_Time) on 15 January 2022. The above animation (credit: Tim Schmit, NOAA/NESDIS/ASPB) shows propogation of the shockwave across the Earth as seen from space (using infrared measurements of water vapor). That shockwave was also measured as changing pressure in weather stations around the world. 

a) The CSV file `data/california_pressure_2022_01_15.csv` contains [Automated Surface Observing System (ASOS)](https://mesonet.agron.iastate.edu/ASOS/) pressure measurements at 64 stations in California taken several hours later that day, observed with 1-minute resolution. Read this data into a `pandas` DataFrame and then plot in the same figure the pressure as a function of time at SBA (Santa Barbara Municipal Airport, in Southern California), OAK (Oakland Airport, Bay Area), and MHS (Mount Shasta City, in Northern California). To allow a better visual comparison between the three, divide each plot by the initial value at the start of each of the timeseries. Overplot a 5-min thick vertical line for each timeseries (with the same color as the line) at the maximum of that series.

```python
import numpy as np
import pandas as pd  # this is by convention
pd.options.display.width = 1000

df = pd.read_csv('data/california_pressure_2022_01_15.csv', index_col = 0, header = 0, parse_dates=True)
df.head(10)
print(len(df.index))
```

```python
subset = ['SBA', 'OAK', 'MHS']

df_sub = df[subset]
df_sub = df_sub/df_sub.mean()
maxtimes = df_sub.idxmax()

ax = df_sub.plot()
for s in subset:
    ax.axvline(maxtimes[s], color = 'k')
```

b) What can you say about the similarity or dissimilarity of the three series? Why argument can you make that the similarities are due to the Tongan shockwave and not other effects? What do you notice about the location of the maximum of the three series?

All three series show a fairly rapid increase to a local maximum in pressure just after 12:00 UTC, followed by a very rapid decline to a local minimum, i.e. a shock wave. Just after the local minimum, the fluctuations in pressure at all three stations seems larger than average. The three series show very little obvious correlation in their behavior more than an hour before or more than a few hours after the shock wave, but pressure changes are relatively smooth.

It is important to note that the shockwave only represents a local maximum, not a global maximum! So the Santa Barbara station's maximum is NOT near the shockwave. This means we need a better metric than maximum to show shockwave timing.

c) On a map of California, create a static vizualization of the timelag of the passage of the shockwave through the 64 stations in California. You might find it helpful to detect significant changes at each station using rolling percent changes on ~5-10 minute intervals (cf. [`.pct_change(periods=...)`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html) in pandas). The file `data/ca_asos.csv` contains the latitude and longitude of the stations. 

```python


#very quick check that rolling pct gives us a good guess
fig, ax = plt.subplots(1,1,figsize=(12,8))
rolling_pct_sub = df_sub.pct_change(periods=10)

ax.plot(df_sub)
ax2=ax.twinx()

ax2.plot(rolling_pct_sub, color='k')
plt.show()

# this shows us what you really want is the maximum negative change because it is more abrupt

rolling_pct = df.pct_change(periods=10)
station_max_time = rolling_pct.idxmin()

# merge the data
station_max_time = station_max_time.to_frame().rename(columns={0: "max_time"})
station_locations = pd.read_csv('data/ca_asos.csv')
station_locations = station_locations.set_index('stid')
station_locations = station_locations.merge(station_max_time, left_index=True, right_index=True, how='right')

station_locations.head(10)
```

```python
# now do the visualization
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib import cm
from datetime import datetime, timedelta

proj = ccrs.PlateCarree()
cmap = cm.hot
fig, ax = plt.subplots(1,1, figsize = (6,10), subplot_kw={'projection':proj})
ax.set_extent([-125, -113, 32, 43], crs=proj)

states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(states_provinces, edgecolor='gray')

time_zero = datetime.strptime('15 January 2022 04:14:45', '%d %B %Y %H:%M:%S')
time_lag = (station_locations['max_time'] - time_zero)
print(time_lag)

im = ax.scatter(station_locations['lon'], station_locations['lat'], transform = proj, c=np.asarray(time_lag)/(1e9*60))
cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04, label = 'Time lag (minutes)')

plt.show()
```

d) Pick one station and determine when you believe the shockwave first passed through that station. Using this information and the time of the explosion determine the speed of the shockwave. How does this compare to the speed of sound in air?

```python
from cartopy import geodesic
lat_0, lon_0 = -20.550, 175.385
station_locations.head(10)
bishop = station_locations.loc['BIH']

gd = geodesic.Geodesic()
dist = gd.inverse((bishop['lon'], bishop['lat']), (lon_0, lat_0))
dist = dist[0,0] # meters
time = time_lag['BIH'].total_seconds()
speed = dist / time
print(f'speed of wave was {speed} m/s')
print('the speed of sound is 331.29 m/s')
print('so they are very similar')
```
