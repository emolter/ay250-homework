## Problem statement
we are given data on the x,y position as a function of time of two moving objects. We know the objects are moving in the same speed and direction the whole time. The data are noisy.  We don't know the noise properties, but can assume the noise on all measurements is the same and that the measurements are independent. 

### Part A
find the posterior of v

```python
import pymc3 as pm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import arviz as az
import pickle
```

```python
# how many CPUs are available for us to multithread?
import multiprocessing
multiprocessing.cpu_count()
```

```python
cut = 100 #None or integer
dset = pd.read_csv('location_data_hw8_2022.csv')
if cut is not None:
    dset = dset[:100]
    filetail = str(cut)
else:
    filetail = ''
print(dset)
```

```python
class velocity_vector:
    
    def __init__(self, x0, y0, v, theta):
        '''
        units of distance dont matter as long as x,y,v are consistent
        theta assumed degrees, 0 is to the right
        '''
        self.x0 = x0
        self.y0 = y0
        self.v = v
        self.theta = theta
        
    def position_at_t(self, t, sigma_x = 0, sigma_y = None):
        '''
        generate noisy data like that given in the hw problem, but for arbitrary t
        sigma is 
        should be able to pass many ts at once
        '''
        if sigma_y is None:
            sigma_y = sigma_x
            
        
        d = self.v*t
        y = self.y0 + d*np.sin(np.deg2rad(self.theta)) + np.random.normal(loc = 0, scale = sigma_y, size = t.shape)
        x = self.x0 + d*np.cos(np.deg2rad(self.theta)) + np.random.normal(loc = 0, scale = sigma_x, size = t.shape)
        
        return x,y
    
    def time_at_given_y(self, yi):
        '''
        solve part B. ignore uncertainty, just find where y = yi given y0, v, theta
        assume t0 = 0. so t can be negative
        '''
        vy = self.v*np.sin(np.deg2rad(self.theta))
        t_at_0 = (yi - self.y0)/vy
        return t_at_0
```

```python
# testing if velocity vector maker works
some_vector = velocity_vector(5, -3, 0.7, -60.)
x, y = some_vector.position_at_t(dset['t'], 30)

plt.scatter(x,y)
plt.show()
```

So we need four parameters in our model to get a complete solution: x0, y0, v, and theta. Note it's important to retain time information. This plot could also be generated using x0,y0 $\approx$ 350, -550 and theta = 120, but that is not the correct answer. However, unless the data were VERY noisy, you could guess that yourself. So this is kinda a good time to set a prior on, e.g., theta to within 90 degrees, or to make the prior on x0, y0 a 2-D Gaussian around the initial point.  But let's not do that for now

```python
# plot the actual data

plt.scatter(dset['red_pos_X'], dset['red_pos_Y'], color = 'red')
plt.scatter(dset['blue_pos_X'], dset['blue_pos_Y'], color = 'blue')
plt.show()
```

```python
dset_color = 'red'
```

```python
# build the model
with pm.Model() as model:
    # set time as immutable data
    t = pm.Data('t', dset['t'])
    
    # Specify priors for x0, y0, v, theta, and uncertainty sigma
    x0 = pm.Uniform('x0', lower=0.5, upper=2) # just uniform for now
    y0 = pm.Uniform('y0', lower=-3, upper=3) # just uniform for now
    v = pm.Exponential('v', lam=1.) # v can take any positive value; use exponential
    theta = pm.VonMises('theta', mu=0., kappa=0.01) # I googled what is good for angles and this popped up
    sigma = pm.Exponential('sigma', lam=1.)

    # now we want our full model. I know how to set this up in an emcee framework but not with pymc3
    # should be normal distributions with sigma unknown
    # and x,y position determined from the four parameters
    x = pm.Normal('x', mu = x0 + v*pm.math.cos(theta)*t, sigma = sigma, observed=dset[f'{dset_color}_pos_X'])
    y = pm.Normal('y', mu = y0 + v*pm.math.sin(theta)*t, sigma = sigma, observed=dset[f'{dset_color}_pos_Y'])
    
```

```python
# run the mcmc and save thinned trace to pickle
with model:
    step = pm.NUTS() 
    trace = pm.sample(10000, step=[step], cores=8, return_inferencedata=False)
    
burned_in = trace[5000:]
thinned = burned_in[::10]
print(thinned)
with open(f'thinned_{dset_color}{filetail}.pickle', 'wb') as handle:
    pickle.dump(thinned, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
## check for good convergence by plotting the trace
with open(f'thinned_{dset_color}{filetail}.pickle', 'rb') as handle:
    thinned = pickle.load(handle)
var_names = ['x0', 'y0', 'v', 'theta', 'sigma']
_ = az.plot_trace(thinned, var_names=var_names)
```

```python
# make a corner plot
import corner
_ = corner.corner(thinned)
```

```python
summary_stats = pm.stats.summary(thinned,  var_names=var_names, round_to = 7) #summary statistics
print(summary_stats)
```

```python
# check if the results are reasonable by plotting the noise-free trajectory of mean values
result_vector = velocity_vector(summary_stats['mean']['x0'], summary_stats['mean']['y0'], summary_stats['mean']['v'], np.rad2deg(summary_stats['mean']['theta']))
xf, yf = result_vector.position_at_t(dset['t'])

plt.scatter(dset[f'{dset_color}_pos_X'], dset[f'{dset_color}_pos_Y'], color = dset_color, label = 'data', alpha = 0.5)
plt.plot(xf, yf, color = 'k', linestyle = '--', label = 'model')
plt.scatter(result_vector.x0, result_vector.y0, marker = '*', s = 200, color = 'k', label = 'starting point')
plt.scatter(dset[f'{dset_color}_pos_X'][0], dset[f'{dset_color}_pos_Y'][0], marker = '*', s = 200, color = f'dark{dset_color}', label = 'first data point')
plt.legend()
plt.show()
```

### Answer A
The mcmc retrieval thinks it can retrieve the velocity very accurately, yielding $v = 2.278 \pm 0.005 \times 10^{-3}$ for the red object and $v = 2.286 \pm 0.011 \times 10^{-3}$ for the blue object ($1\sigma$ errors). See the corner plot above for distributions. To toggle red or blue here, see the dset\_color variable in the cell where the model is built. The results look reasonable based on the plot of model vs data. Note that I assumed the uncertainty in the x-direction is independent of the uncertainty in the y-direction, but has the same rms value.

### Part B 
figure out the first time (with error bars) at which $Y_{blue} \geq 0$ and $Y_{red} \leq 0$

```python
# load the traces back in
with open(f'thinned_red{filetail}.pickle', 'rb') as handle:
    thinned_red = pickle.load(handle)
with open(f'thinned_blue{filetail}.pickle', 'rb') as handle:
    thinned_blue = pickle.load(handle)
```

```python
# calculate time that sign of y switches for every draw. this will give uncertainty
vectors_red = velocity_vector(thinned_red[:]['x0'], thinned_red[:]['y0'], thinned_red[:]['v'], np.rad2deg(thinned_red[:]['theta']))
t_at_0_red = vectors_red.time_at_given_y(0.0)
vectors_blue = velocity_vector(thinned_blue[:]['x0'], thinned_blue[:]['y0'], thinned_blue[:]['v'], np.rad2deg(thinned_blue[:]['theta']))
t_at_0_blue = vectors_blue.time_at_given_y(0.0)

plt.hist(t_at_0_red, bins = 20, color = 'red', alpha = 0.5)
plt.hist(t_at_0_blue, bins = 20, color = 'blue', alpha = 0.5)
plt.xlabel('t at y=0')
plt.ylabel('number of draws')
plt.show()
```

```python
[t5, t16, t50, t84, t95] = np.percentile(t_at_0_blue, [0.05, 16, 50, 84, 95])
plt.hist(t_at_0_blue, bins = 20, color = 'blue', alpha = 0.5)
plt.show()
print(t50, t50 - t5, t95 - t50)
```
